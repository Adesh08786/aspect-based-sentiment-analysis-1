```python
import xml.etree.ElementTree as ET
import nltk

def get_list(path):
    tree=ET.parse(path)
    root = tree.getroot()
    text_list = []
    id_list=[]
    new_list=[]
    opinion_list = []
    for review in root.findall('Review'):
        text_string=""
        opinion_inner_list=[]
        for sent in review.findall('./sentences/sentence'):
            ids=sent.attrib['id']
            id_list.append(ids)
            text_string= text_string+ " "+ sent.find('text').text
            new_list.append(ids)
            new_list.append(sent.find('text').text)
        
        text_list.append(text_string)
        for opinion in review.findall('./Opinions/Opinion'):
            opinion_dict = {
                opinion.get('category').replace('#','_'): opinion.get('polarity')
            }
            opinion_inner_list.append(opinion_dict)
        opinion_list.append(opinion_inner_list)
    return text_list,opinion_list,new_list
train_text_list,train_opinion_list,new_list = get_list("Downloads/ABSA16_Laptops_Train_English_SB2.xml")

```

```python
def detect_terms(temp,sentence,nnlist):
    aspect_term = []
    for relation in temp:
        if relation[0]=='amod':
            aspect_term.append(relation[2].split('-')[0])
        if (relation[0]=='acomp' or relation[0]=='xcomp' or relation[0]=='nmod' or relation[0]=='dobj'
        or relation[0]=='nsubj' or relation[0]=='nsubjpass' or relation[0]=='xcomp' 
        or relation[0]=='pobj' or relation[0]=='abbrev'):
            aspect_term.append(relation[4].split('-')[0])
    # print aspect_term
    aspect_term = list(set(aspect_term))
    for term in aspect_term:
        if term not in nnlist:
            aspect_term.remove(term)
    for relation in temp:
        if relation[0]=='conj' and relation[2].split('-')[0] in aspect_term:
            aspect_term.append(relation[4].split('-')[0])
    aspect_term = list(set(aspect_term))
    return aspect_term
```
